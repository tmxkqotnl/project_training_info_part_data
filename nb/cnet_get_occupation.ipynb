```python
jsonurl = ''.join(['https://www.career.go.kr/cnet/openapi/getOpenApi?apiKey=', os.getenv('CAREER_NET_API_KEY')])
jsonurl = '&'.join([jsonurl,'svcType=api','svcCode=JOB','contentType=json','gubun=job_dic_list','thisPage=2','perPage=454'])
```

```python
lst = []
for i in range(1,50):
    jsonurl = ''.join(['https://www.career.go.kr/cnet/openapi/getOpenApi?apiKey=', os.getenv('CAREER_NET_API_KEY')])
    jsonurl = '&'.join([jsonurl,'svcType=api','svcCode=JOB','contentType=json','gubun=job_dic_list','thisPage='+str(i),'perPage=20'])
    response = requests.get(jsonurl)
    json = response.json()
    lst.extend(json['dataSearch']['content'])

import pandas as pd

job_dic = pd.json_normalize(lst)
job_dic
```
