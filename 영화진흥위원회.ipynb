```python
import requests
import json
import datetime
import pandas as pd

def get_movieInfo(movieCd):
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key=130ad25d8935a673ca47beb43436aa85&movieCd='+str(movieCd)
    res = requests.get(url)
    text = res.text
    #print(text)

    d = json.loads(text)

    movieInfo = d['movieInfoResult']['movieInfo']

    return movieInfo

current_datetime = datetime.datetime(2018, 2, 20)
l = []
for i in range(30):
    targetDt = current_datetime.strftime('%Y%m%d') #20180220
    #print(i)
    #print(targetDt)
    #url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=130ad25d8935a673ca47beb43436aa85&targetDt=20180220'
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=130ad25d8935a673ca47beb43436aa85&targetDt='+targetDt
    res = requests.get(url)
    text = res.text
    #print(text)

    d = json.loads(text) 

    for b in d['boxOfficeResult']['dailyBoxOfficeList']:
        movieInfo = get_movieInfo(b['movieCd'])

        print(targetDt,
              b['rank'],
              b['rankOldAndNew'],
              b['movieCd'],
              b['movieNm'],
              b['salesAmt'],
              b['audiCnt'],
              #
              movieInfo['showTm'],
              len(movieInfo['actors']),
              len(movieInfo['showTypes']))
        l.append([targetDt,
              b['rank'],
              b['rankOldAndNew'],
              b['movieCd'],
              b['movieNm'],
              b['salesAmt'],
              b['audiCnt'],
              #
              movieInfo['showTm'],
              len(movieInfo['actors']),
              len(movieInfo['showTypes'])])

    current_datetime = current_datetime + datetime.timedelta(days=1)

df = pd.DataFrame(l, columns=['targetDt',
                              'rank',
                              'rankOldAndNew',
                              'movieCd',
                              'movieNm',
                              'salesAmt',
                              'audiCnt',
                              #
                              'showTm',
                              'actors',
                              'showTypes'])
# df.to_excel('kobis.xlsx', index=False)
```
